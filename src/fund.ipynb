```python
# group by 
df[["operate_mode", "id"]].groupby("operate_mode").count().rename({"id": "count"})
```

```python
df[df['operate_mode'] == "ETF"][["underlying_asset_type", "id"]].groupby("underlying_asset_type").count()
```

```python


etf_df = df[df['operate_mode'] == "ETF"] 
etf_df = etf_df[etf_df["underlying_asset_type"] == "股票型"]
etf_df = etf_df.sort_values("list_date", ascending=False)
etf_df.to_excel("etf.xlsx")
```
